```python
df_country = pd.read_excel('C:/Users/Satyam/Downloads/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# Joining Datasets/ Tables

final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head(5)
```

```python
# Find the number of times particular country present in dataset

final_df.Country.value_counts()
```

```python
#Plot Pie Chart for Countries

country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val, labels = country_names)
```

Country names ovelapped. To avoid this take onlt top 3 countries

```python
plt.pie(country_val[:3], labels = country_names[:3], autopct='%1.2f%%')
```

Observation:

    1. Maximum records/ transactions in dataset is from India. Then followed by USA and UK respectively.
    2. Maximum business of Zomato is concentrated in India.

```python
# Lets analyse ratings from dataset
# We have used .rename to .rename '0' and .reset_index() to convert data into dataframe


ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

1.Observation:

    1. When Rating is between 4.5 to 4.9, rating is Excellent. 
    2. When Rating is between 4.0 to 4.4, rating is Very Good.
    3. When Rating is between 3.5 to 3.9, rating is Good.
    4. When Rating is between 2.5 to 3.4, rating is Average.
    5. When Rating is between 2.4 to 1.8, rating is Poor.
    6. When Rating is between 0, rating is not Given.

2.Observation: Most of the people has not given ratings.

```python
# Lets plot ratings count for each rating using seaborn

plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
# Plot batplot with color as per 'Rating Color' using 'palette' function.

sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

Observation: 

    1. Not rated count is very high.
    2. Maximum number of ratings are between 2.5 to 3.4

```python
# Plot countplot of categorical variable 'Rating color'

sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

```python
# Find the country names those have given 0 rating.

final_df[final_df['Aggregate rating'] == 0.0].groupby('Country').size().reset_index()
```

Obervation:
